```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
movie_file = "Data/movies_metadata.csv"
movie_df = pd.read_csv(movie_file)
movie_df.head() 
```

```python
tmdb_file = "Data/tmdb_5000_movies.csv"
tmdb_df = pd.read_csv(tmdb_file)
tmdb_df.head() 
```

```python
# Create a filtered dataframe from specific columns
movies_cols = [ "imdb_id","title"]
movies_df= movie_df[movies_cols].copy()


# Clean the data by dropping duplicates and setting the index
movies_df.drop_duplicates("title", inplace=True)
movies_df.set_index("title", inplace=True)

movies_df.head()
```

```python
# Create a filtered dataframe from specific columns
tmdbs_cols = ["budget","popularity","production_companies","revenue","title"]
tmdbs_df= tmdb_df[tmdbs_cols].copy()


# Clean the data by dropping duplicates and setting the index
tmdbs_df.drop_duplicates("title", inplace=True)
tmdbs_df.set_index("title", inplace=True)

tmdbs_df.head()
```

```python
rating_file = "Data/ratings_small.csv"
rating_df = pd.read_csv(rating_file)
rating_df.head() 
```

```python
link_file = "Data/links_small.csv"
link_df = pd.read_csv(link_file)
link_df.head() 
```

```python
# Create a filtered dataframe from specific columns
ratings_cols = ["movieId","rating"]
ratings_df= rating_df[ratings_cols].copy()

ratings_df.set_index("movieId", inplace=True)

ratings_df.head()
```

```python
# Create a filtered dataframe from specific columns
links_cols = ["movieId","imdbId"]
links_df= link_df[links_cols].copy()

links_df.set_index("movieId", inplace=True)

links_df.head()
```

```python
merged_df = pd.merge(movies_df, tmdbs_df, on='title', how='inner')
merged_df
```

```python
combined_df = pd.merge(ratings_df, links_df, on='movieId', how='inner')
combined_df
```
